# Missing Values (Categorical)


```python
#Let us capture all the nan values
#Firt lets handle categorical features which are missing.

cat_features_nan = [feature for feature in dataset.columns if dataset[feature].isnull().sum() > 1 and dataset[feature].dtypes == 'O']
```

```python
cat_features_nan
```

```python
dataset[cat_features_nan].isnull().sum()
```

```python
for feature in cat_features_nan:
    print("{}: {}% missing values".format(feature, np.round(dataset[feature].isnull().mean(), 4)))
```

```python
#Now we have figure out the categorical features which has null values now lets replace the null values and with new label "Missing".


def create_new_label(dataset, cat_features_nan):
    data = dataset.copy()
    data[cat_features_nan] = data[cat_features_nan].fillna('Missing')
    return data

dataset= create_new_label(dataset, cat_features_nan)
```

```python
dataset[cat_features_nan].isnull().sum()
```

# Missing values (Numerical values)

```python
numerical_nan_feature = [feature for feature in dataset.columns if dataset[feature].isnull().sum() > 1 and dataset[feature].dtypes != 'O']
```

```python
numerical_nan_feature
```

```python
#Lets represent into percentage
for feature in numerical_nan_feature:
    print("{}: {} % Missing values".format(feature, np.around(dataset[feature].isnull().mean(), 4)))
```

```python
for feature in numerical_nan_feature:
    # As we already know that there are outliers in numerical features so will replacie nan with median value of the particular feature dataset
    median_value = dataset[feature].median()
    
    #To get more information from the data we will create one more feature which will have missing values as "1" and "0" as value exsist
    dataset[feature+ '_nan'] = np.where(dataset[feature].isnull(),1,0)
    dataset[feature].fillna(median_value, inplace=True)
```

```python
#After comuting missing values
for feature in numerical_nan_feature:
    print("{}: {} % Missing values".format(feature, np.around(dataset[feature].isnull().mean(), 4)))
```

```python
dataset[['LotFrontage_nan','LotFrontage']].head(20)
```

```python
#Note: 1 represent missing value and 0 represent not null
#we also can see that median value corresponding to 1.
```

```python
#Date time varialble (Temporal variables) Date Time variables.

date_features = [feature for feature in dataset.columns if 'Year' in feature or 'Yr' in feature]
```

```python
date_features= date_features[:4]
date_features
```

```python
for feature in date_features:
    if feature != 'YrSold':
        dataset[feature] = dataset['YrSold'] - dataset[feature]
```

```python
dataset[date_features]
```

```python
#In exploratory data analysis we come to know that the numerical features are skewed, so now we will perform log normal distribution.

numerical_continues_feature = ['LotFrontage', 'LotArea', '1stFlrSF','GrLivArea', 'SalePrice']
```

```python
numerical_continues_feature
```

```python
for feature in numerical_continues_feature:
    dataset[feature] = np.log(dataset[feature])
```

```python
dataset.head()
```

# Handling Rare categorical feature

```python
#if Occurence of particular categorical feature value is less then 1 % then we can remove that categorical value from the dataset
categorical_feature = [feature for feature in dataset.columns if dataset[feature].dtype == 'O']
categorical_feature
```

```python
for feature in categorical_feature:
    temp = dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df = temp[temp>0.01].index
    dataset[feature] = np.where(dataset[feature].isin(temp_df), dataset[feature], 'Rare_var')
```

```python
dataset.head(100)
```

```python

for feature in categorical_feature:
    labels_ordered=dataset.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]=dataset[feature].map(labels_ordered)
```

```python
#temp_df mostly used categorical values for one of any particular feature. 
```

# Feature Scaling

```python
#Before scalling
dataset.head()
```

```python
#On which features we must do feature scaling?
# scaling data between 0 to 1
#After Scaling
feature_scal = [feature for feature in dataset.columns if feature not in ['Id','SalePrice']]
feature_scal

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(dataset[feature_scal])
```

```python
scaler.transform(dataset[feature_scal])
```

```python
new_data = pd.concat([dataset[['Id','SalePrice']].reset_index(drop=True),
                   pd.DataFrame(scaler.transform(dataset[feature_scal]), columns=feature_scal)],axis=1)
```

```python
new_data.head()
```
